```python
import time #To know the time and date 
fd=open('produt.txt','a')
fd.close()    #To create a file 
name=input("can i know your name please: ")   #Taking inputs from the user to maintain sales list
email=input("what is your email id     : ")
phone=input("what is your phone number : ")
item_list=[]
print("what do you want buy")
print("please look into the menu ")       #showing menu to the customer 
anymore="Y"
Total_bill=0
    
def selling_product(serialno,quantity):    #function for selling products to the customers
    fd=open('produt.txt','r')
    product=fd.read()
    fd.close()                             #opening the file in a read mode 
    product=product.split('\n')            #splitting the string in the file to list by using \n
    l=[]
    for i in product:
        m=[]
        pdsp=i.split(',')                  #splitting each list string element into list by using ,

        if(int(serialno)==int(pdsp[0])):   #condition to check whether required element is present in the store
            if(int(pdsp[3])>=int(quantity)): #condition to check whether reqired quantity of element is present or not
                m.append(pdsp[1])
                amount=int(quantity)*int(pdsp[2])   #calculating price of the product bought
                m.append(pdsp[2])
                m.append(quantity)
                pdsp[3]=str(int(pdsp[3])-int(quantity))   #updating the value of quantity in the list
                m.append(amount)
                item_list.append(m)
            else:
                print("the item you are looking for is available")     #this condition is for quantity insuffesience
                print("but we doesnot have that much quantity")        
                print("we have only this much quantity : ",pdsp[3])
                print("if you require them please select y or else n")  #conforming from the customer whether they want to buy the quantity present in the store
                ip=input()
                if(ip=="y"  or ip=="Y"):
                    amount=int(pdsp[3])*int(pdsp[2])
                    m.append(pdsp[1])
                    m.append(pdsp[2])
                    m.append(pdsp[3])
                    m.append(amount)
                    item_list.append(m)
                    pdsp[3]="0"
                else:
                    print("sorry for the inconvience")
    
        l.append(pdsp)

    fd=open('produt.txt','w')
    for i in range(len(l)):                                   #updating the file after the product selled
        if(i==(len(l)-1)):
            fd.write(l[i][0]+","+l[i][1]+","+l[i][2]+","+l[i][3])
        else:
            fd.write(l[i][0]+","+l[i][1]+","+l[i][2]+","+l[i][3]+"\n")

    fd.close()

while(anymore=="Y" or anymore=="y"):
    
    serialno=input("enter the serial number: ")              #obtaining serial number,quantity of the product they want
    quantity=input("enter the quantity     : ")
    selling_product(serialno,quantity)                       #calling function
    anymore=input("Do you want to buy anything more if you require print y or else n :")  #checking whether they want buy more products or not 

if(len(item_list)>0):    #generating the bill for the customers
    print("-----------------------------------------------------------------------------------------------------------------")
    print("Name of the product"+"           "+"cost of the item"+"     "+"quantity bought"+"    "+"totalprice")
    for i in range(len(item_list)):
        Total_bill=Total_bill+item_list[i][3]
        print(item_list[i][0]+" "*(33-len(item_list[i][0]))+item_list[i][1]+" "*(20-len(item_list[i][1]))+item_list[i][2]+" "*(21-len(item_list[i][2]))+str(item_list[i][3]))
    print("------------------------------------------------------------------------------------------------------------------")
    print("total bill"+" "*64+str(Total_bill)+"(Rupees)")
    print("------------------------------------------------------------------------------------------------------------------")
    fd=open('sales.txt','a')
    fd.write("Name of the customer: "+name+","+"email address of the customer: "+email+","+"phone number of the customer: "+phone+","+"Total Bill: "+str(Total_bill)+","+"date and time: "+time.ctime()+"\n")
    fd.close()
else:
    print("Sorry sir/madam please visit us again")   
    

    
```
